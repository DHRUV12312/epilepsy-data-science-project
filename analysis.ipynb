```python
import pandas as pd

data = pd.read_csv("data/EEG_Signal.csv")

print(data.head())
print(data.shape)
```

```python
data["Labels"].value_counts()
```

```python
data = data.sample(5000)
```

```python
import matplotlib.pyplot as plt

plt.plot(data["Signal"][:200])
plt.title("Sample EEG Brain Signal")
plt.xlabel("Time")
plt.ylabel("Signal Value")
plt.show()
```

```python
import seaborn as sns

sns.countplot(x="Labels", data=data)
plt.title("Seizure vs Non-Seizure")
plt.show()
```
